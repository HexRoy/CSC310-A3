#### Imports

```python
import pandas as pd
from sklearn import tree
from treeviz import tree_print
from sklearn.metrics import accuracy_score
```

# Tennis Dataset

#### Creates the dataframe

```python
tennis_df = pd.read_csv('tennis_numeric.csv')
```

```python
tennis_df
```

#### Creates the entropy tree

```python
tennis_dtree = tree.DecisionTreeClassifier(criterion='entropy')
```

### Split the data into numerical and categorical

#### Categorical

```python
tennis_categorical = pd.DataFrame(tennis_df['play'])
```

```python
tennis_categorical.head()
```

#### Numerical

```python
tennis_numerical = tennis_df.drop(['play'], axis=1)
```

```python
tennis_numerical.head()
```

#### Create the entropy tree


```python
tennis_dtree = tree.DecisionTreeClassifier(criterion='entropy')
```

#### Build the model


```python
tennis_dtree.fit(tennis_numerical, tennis_categorical)
```

#### Print the tree

```python
tree_print(tennis_dtree, tennis_numerical)
```

#### Produces the array of labels

```python
predict_array = tennis_dtree.predict(tennis_numerical)
```

```python
predicted_labels = pd.DataFrame(predict_array)
```

```python
predicted_labels.columns = ['play']
```

#### A list of the predicted labels 

```python
predicted_labels.head()
```

#### A list of the actual labels

```python
tennis_categorical.head()
```

#### Checks to see if the predicted label and target label are the same

```python
predicted_labels.equals(tennis_categorical)
```

### Model Accuracy

```python
print("Our model accuracy is: {}".format(accuracy_score(tennis_categorical, predicted_labels)))
```

```python
dtree2 = tree.DecisionTreeClassifier(criterion='entropy', max_depth=2)
dtree2.fit(tennis_numerical,tennis_categorical)
```

```python
tree_print(dtree2, tennis_numerical)
```

```python
predict_array2 = dtree2.predict(tennis_numerical)      
```

```python
predicted_labels2 = pd.DataFrame(predict_array2) 
```

```python
predicted_labels2.columns = ['play']
```

```python
print("Our model accuracy is: {}".format(accuracy_score(tennis_categorical, predicted_labels2)))
```
